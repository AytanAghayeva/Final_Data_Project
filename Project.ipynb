```python
#Remove duplicate rows
df = df.drop_duplicates()
#df = df.drop_duplicates(subset=['employee_residence'], keep='first', inplace=True) - based on specific columns
```

```python
#Renaming columns
df.rename(columns={'work_year': 'Work_Year'}, inplace=True)
df.rename(columns={'job_title': 'Job_Title'}, inplace=True)
df.rename(columns={'job_category': 'Job_Category'}, inplace=True)
df.rename(columns={'salary': 'Salary'}, inplace=True)
df.rename(columns={'employee_residence': 'Employee_Residence'}, inplace=True)
df.rename(columns={'experience_level': 'Experience_Level'}, inplace=True)
df.rename(columns={'employment_type': 'Employment_Type'}, inplace=True)
df.rename(columns={'work_setting': 'Work_Setting'}, inplace=True)
df.rename(columns={'company_location': 'Company_Location'}, inplace=True)
df.rename(columns={'company_size': 'Company_Size'}, inplace=True)
```

```python
#Sorting by multiple columns
df.sort_values(by=['Job_Category','Experience_Level'], ascending=[True, True], inplace=True)
```

```python
df
```

```python
df_grouped = df.groupby('Employee_Residence').agg({'Salary' : 'sum'})
df_grouped
```

```python
df_grouped_sorted = df_grouped.sort_values(by='Salary', ascending=False)
df_grouped_sorted
```

```python
df['Company_Size'].unique()
```

```python
size_mapping = {'S': 'Small', 'M': 'Medium', 'L': 'Large'}
df['Company_Size'] = df['Company_Size'].map(size_mapping)
```

```python
df['Experience_Level'].unique()
```

```python
experience_mapping = { 'Executive': 'Executive', 'Mddl': 'Middle', 'Snr': 'Senior', 'Entry': 'Entry'}
df['Experience_Level'] = df['Experience_Level'].map(experience_mapping)           
```

```python
df[['Experience_Level']].head()
```

```python
print(df['Salary'])
```

```python
'''If one person in the salary data earns 1 million, the average (mean) will go up a lot.
But the median is just the middle number, so it is not affected by outliers.'''
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)
```

```python
print(df['Salary'])
```

```python
print(df['Work_Year'])
```

```python
#The best way is to fill missing values with the mode (the most common value).
df['Work_Year'] = df['Work_Year'].fillna(df['Work_Year'].mode()[0])
```

```python
print(df['Work_Year'])
```
